```python
c=5
H=10 #km

#Define the function f(x) as executable 
def f(z):
    f = z*sy.exp(-z/H)
    return f

#Base con condiciones Mixtas normalizada; componente z
varphiM_k=sy.sqrt(2/c)*sy.cos(((l+(1/2))*sy.pi*z)/c)

#Base con condiciones Neumann normalizada; componente z
varphiN_k=sy.sqrt(2/c)*sy.cos((l*sy.pi*z)/c)

#------------------Cálculo del coeficiente fk  ----------------
#-----------considerando condiciones Mixtas y Neumann----------
w1=(l*sy.pi/c)
w2=((1/2)+l)*(sy.pi/c)
gama=H/(1+(w2*H)**2)

#print('\n\n Coeficiente de la serie f_k con condiciones de frontera Mixtas \n')

fk1_M=sy.sqrt(2/c)*gama*(((-1)**l)*c*H*w2*sy.exp(-c/H)+gama*(1-(H*w2)**2))

#sy.pprint(fk1_M)

serieM_fl = 0
for k in range(0,20):
    serieM_fl = serieM_fl + ((varphiM_k*fk1_M).subs(l,k))
sy.plot(serieM_fl,(z,-6,6))

#--------------------------------------------------

#print('\n\n Coeficiente de la serie f_k con condiciones de frontera Neumann \n')

fk1_N=((gama)**2)*sy.sqrt(2/c)*(((-1)**l)*sy.exp(-c/H)*((c/gama)-1+((H*w1)**2))+1-((H*w1)**2))

#sy.pprint(fk1_N)

f0 = sy.integrate((1/c)*f(z),(z,0,c))
serieN_fl = f0
for k in range(1,20):
    serieN_fl = serieN_fl + ((varphiN_k*fk1_N).subs(l,k))
sy.plot(serieN_fl,(z,-6,6))
```

```python
#----------------------Norma de f(z) -----------------------------------

fc = sy.integrate(f(z)**2,(z,0,c))
normaf = np.sqrt(float(fc))
print( 'La norma de f(z)=',f(z),' en el intervalo [ 0 ,',c,'] es: ', normaf)


#----------------------Norma de f_l---Dirichlet--------------------------------

print('\n\n Condiciones de frontera Mixtas')

norma_fl=0
for k in range(0,20): 
    norma_fl = norma_fl + ((fk1_M)**2).subs(l,k)
print("La norma de la serie f_l(z) con l=200 es: ", float(sy.sqrt(norma_fl)))  

#----------------------Norma de f_l---Neumann--------------------------------

print('\n\n Condiciones de frontera Neumann')

f0 = sy.integrate(sy.sqrt(1/c)*f(z),(z,0,c))
norma_fl=(f0)**2
for k in range(1,20): 
    norma_fl = norma_fl + ((fk1_N)**2).subs(l,k)
print("La norma de la serie f_l(z) con l=200 es: ", float(sy.sqrt(norma_fl))) 
```

```python
#------Cálculo del flujo y porcentaje de masa BVP1-----------

beta=0.005077181196118556

def F(x,y,z):
    F = g(x)*e(y)*f(z)
    return F

#-...-Límites de integración e inicialización-------------------
xyM = 12
xym = 8
zM = 3
zm = 2

#-----------Calclulo de la integral de Fmnl2-----------------
for n in range(3):
    int_ser1=sy.integrate(serieN_gm,(x,xym,xyM))
    int_ser2=sy.integrate(serieN_en,(y,xym,xyM))
    int_ser3=sy.integrate(serieM_fl,(z,zm,zM))
    int_Fmnl1 = int_ser1*int_ser2*int_ser3
    
#-------------Calclulo de la integral de F2------------------
    int_F1 = sy.integrate(F(x,y,z),(x,xym,xyM),(y,xym,xyM),(z,zm,zM))
    flux_1 = beta*(int_F1-int_Fmnl1)
    
#------- ---Porcentaje de Masa que fluye en 3hr---------------

    PorcentajeMasa=(3*100*3600*abs(flux_1))/((xyM-xym)*(xyM-xym)*(zM-zm))
    
#---------------------Resultados------------------------------
    print('\n\n Flujo en la subregión (',xym,',',xyM,')^{2}x(',zm,',',zM,') --> ',float(flux_1))
    print(' %Masa que fluye en la subregión en 3hr   --> ',float(PorcentajeMasa))
    xyM = xyM + 4
    xym = xym - 4
    zM = zM + 1
    zm = zm - 1
```
